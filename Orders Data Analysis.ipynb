```python
# Converting unknown and Not Available values of Ship Mode column to NaN values

df=pd.read_csv("orders.csv",na_values=["unknown","Not Available"])
df["Ship Mode"].unique()
```

```python
# Renaming Columns

df.columns = df.columns.str.lower()
df[:4]
```

```python
df.columns=df.columns.str.replace(" ","_")
df[:4]
```

```python
df["order_date"]=pd.to_datetime(df["order_date"])
```

```python
df.info()
```

```python
# Deriving New columns - Discount, Sale Price and Profit
```

```python
df[:2]
```

```python
df["discount"]=(df.discount_percent/100)*df.list_price
df[:2]
```

```python
df["sale_price"]=df.list_price-df.discount
df["profit"]=df["sale_price"]-df["cost_price"]
df[:2]
```

```python
df[:4]
```

```python
# Dropping columns - Cost Price, List Price, Discount Percent

df.drop(["cost_price","list_price","discount_percent"],axis=1,inplace=True)
```
